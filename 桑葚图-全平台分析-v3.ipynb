```python
import pandas as pd
import numpy as np
from pyecharts.charts import Sankey
from pyecharts import options as opts
from pyecharts.globals import CurrentConfig, NotebookType, OnlineHostType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_NOTEBOOK
```

```python
# 统计日期
date = '0528'
```

```python
def get_node_links(df):
    nodes = []
    for i in range(2):
        values = df.iloc[:, i].unique()
        for value in values:
            dic = {}
            dic['name'] = value
            nodes.append(dic)
            
    links = []
    for i in df.values:
        dic = {}
        dic['source'] = i[0]
        dic['target'] = i[1]
        dic['value'] = i[2]
        links.append(dic)
    return nodes, links 
```

```python
# GMV国家分布桑葚图


path1 = "data/sankey_" + str(date) + "_01.csv"


df = pd.read_csv(path1)

var_list = ['store_country','receiver_country_name','gmv']
df1 = df[var_list].rename(columns={"store_country": "source", "receiver_country_name":"target", "gmv":"value"})
df2 = df1[~(df1.source == "中国_店铺")]
df11 = df1[(df1.source == "中国_店铺")]
df1.head()


# 排序
df11['rk'] = df11.groupby(['source'])['value'].rank(method='first', ascending=False)
df2['rk'] = df2.groupby(['source'])['value'].rank(method='first', ascending=False)


#  每个国家只取top10 且gmv超过1000美金的记录
df1_2 = df11[(df11.rk <= 20)&(df11.value >= 1000)]
df2_2 = df2[(df2.rk <= 10)&(df2.value >= 1000)]



nodes_1, links_1 = get_node_links(df1_2)
nodes_2, links_2 = get_node_links(df2_2)
```

```python
pic1 = (
    Sankey()
    .add('',
        nodes=nodes_1,
        links=links_1,
        linestyle_opt=opts.LineStyleOpts(opacity=0.3, curve=0.5, color="source"),
        label_opts=opts.LabelOpts(position="right"),
        node_gap=15,
#         orient="vertical",
    )
    .set_global_opts(title_opts=opts.TitleOpts(title='近1个月订单店铺注册国家(中国)与收货国家分布-2024'+str(date)))
)
pic1.render("sankey_" + str(date) + "_01.html")
pic1.render_notebook()
```

```python
pic2 = (
    Sankey()
    .add('',
        nodes=nodes_2,
        links=links_2,
         node_align='justify',
        node_gap=15,
        linestyle_opt=opts.LineStyleOpts(opacity=0.3, curve=0.5, color="source"),
        label_opts=opts.LabelOpts(position="right"),
#         orient="vertical",
    )
    .set_global_opts(title_opts=opts.TitleOpts(title='近1个月订单店铺注册国家(剔除中国)与收货国家分布-2024' + str(date)))
)
pic2.render("sankey_" + str(date) + "_02.html")
pic2.render_notebook()
```

```python
# 支付方式分布桑葚图


path2 = "data/sankey_" + str(date) + "_02.csv"

df = pd.read_csv(path2)
# 变量字段替换
var_list = ['receiver_country_name','pay_method','gmv']
df1 = df[var_list].rename(columns={"receiver_country_name": "source", "pay_method":"target", "gmv":"value"})
df2 = df1[~(df1.source == "美国")]
df11 = df1[(df1.source == "美国")]

# 排序
df11['rk'] = df11.groupby(['source'])['value'].rank(method='first', ascending=False)
df2['rk'] = df2.groupby(['source'])['value'].rank(method='first', ascending=False)

#  每个国家只取top10 且gmv超过1000美金的记录
df1_2 = df11[(df11.rk <= 20)&(df11.value >= 1000)]
df2_2 = df2[(df2.rk <= 10)&(df2.value >= 1000)]


nodes_3, links_3 = get_node_links(df1_2)
nodes_4, links_4 = get_node_links(df2_2)
```

```python
pic3 = (
    Sankey()
    .add('',
        nodes=nodes_3,
        links=links_3,
        linestyle_opt=opts.LineStyleOpts(opacity=0.3, curve=0.5, color="source"),
        label_opts=opts.LabelOpts(position="right"),
        node_gap=15,
#         orient="vertical",
    )
    .set_global_opts(title_opts=opts.TitleOpts(title='近1个月订单收货地(美国)与支付方式分布top20-2024' + str(date)))
)
pic3.render("sankey_" + str(date) +"_03.html")
pic3.render_notebook()
```
