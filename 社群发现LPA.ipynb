```python
import matplotlib.pyplot as plt
import networkx as nx
from networkx.algorithms.community import asyn_lpa_communities as lpa
```

```python
# 空手道俱乐部数据
G = nx.karate_club_graph()
com = list(lpa(G))
print("社区数量", len(com))
```

```python
com
```

```python
# 画图
pos = nx.spring_layout(G)
NodeId = list(G.nodes())
node_size = [G.degree(i)**1.2*90 for i in NodeId]

plt.figure(figsize = (8,6)) # 设置图片大小
nx.draw(G,pos, 
        with_labels=True, 
        node_size =node_size, 
        node_color='w', 
        node_shape = '.'
       )

'''
node_size表示节点大小
node_color表示节点颜色
with_labels=True表示节点是否带标签
'''
color_list = ['pink','orange','r','g','b','y','m','gray','black','c','brown']

for i in range(len(com)):
    nx.draw_networkx_nodes(G, pos, 
                           nodelist=com[i], 
                           node_color = color_list[i+2],  
                           label=True)
plt.show()

```
